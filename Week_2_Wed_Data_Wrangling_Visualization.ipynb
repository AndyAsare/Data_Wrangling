```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender_Info', hue='Gender_Info',data=df, palette='viridis')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
legend_labels = ['Female', 'Male']
plt.legend(labels=legend_labels)
plt.show()
```

```python
plt.figure(figsize=(7, 3))
sns.boxplot(x='Final_Grade',data=df)
plt.title('Boxplot of Final Grades')
plt.xlabel('Final Grade')
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.histplot(x='Final_Grade', data=df)
plt.title('Hisplot of Final Grades')
plt.xlabel('Final Grade')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='StudyHours_PerWeek', y='Final_Grade', data=df)
plt.title('Scatter Plot of Study Hours vs. Final Grade')
plt.xlabel('Study Hours Per Week')
plt.ylabel('Final Grade')
plt.show()
```

```python
# prompt: generate a lineplot for four variable on Final Grade

# Select four variables for the lineplot
variables = ['Attendance_Rate', 'Previous_Grade']

# Create a figure and axes
plt.figure(figsize=(12, 6))

# Loop through the selected variables
for var in variables:
  # Create a Scatterplot for each variable against Final_Grade
  sns.scatterplot(x=var, y='Final_Grade', data=df, label=var)

# Set the title and labels
plt.title('Scatterplot of Final Grade against Selected Variables')
plt.xlabel('Variable Values')
plt.ylabel('Final Grade')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Student_Name', y='Final_Grade', hue='Parental_Support', data=df, palette=('viridis'))
plt.title('Final Grade by Student Name and Parental Support')
plt.xlabel('Student Name')
plt.ylabel('Final Grade')
plt.xticks(rotation=35, ha='right')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the sum of final grades for each name
name_grades = df.groupby('Student_Name')['Final_Grade'].sum()

# Create the pie chart
plt.figure(figsize=(7, 7))
plt.pie(name_grades, labels=name_grades.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
plt.title('Doughnut Chart of Names and Final Grades')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender_Info', y='Final_Grade', hue='Gender_Info', data=df, palette=('viridis'))
plt.title('Final Grade Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Final Grade')
plt.show()
```

```python
average_final_grade_by_gender = df.groupby('Gender_Info')['Final_Grade'].mean()
print(average_final_grade_by_gender)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix, excluding non-numeric columns
correlation_matrix = df.select_dtypes(include=['number']).corr()

# Create the heatmap
plt.figure(figsize=(10, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# prompt: take out schoolID from the correlation heatmap

# Calculate the correlation matrix, excluding non-numeric columns and 'SchoolID'
correlation_matrix = df.drop('StudentID', axis=1).select_dtypes(include=['number']).corr()

# Create the heatmap
plt.figure(figsize=(8, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# prompt: visualize relationships between Attendance_Rate and final grades variables with P value score and R-Square score

import statsmodels.formula.api as smf
import scipy.stats as stats

# Create a scatter plot
plt.figure(figsize=(8, 4))
sns.lineplot(x='Attendance_Rate', y='Final_Grade', data=df)
plt.title('Scatter Plot of Attendance Rate vs. Final Grade')
plt.xlabel('Attendance Rate')
plt.ylabel('Final Grade')
plt.show()

# Calculate the correlation between 'StudyHours_PerWeek' and 'Final_Grade'
correlation = df['StudyHours_PerWeek'].corr(df['Final_Grade'])

# Print the correlation score
print(f"Correlation between Study Hours Per Week and Final Grade: {correlation:.2f}")

# Fit a linear regression model
model = smf.ols('Final_Grade ~ Attendance_Rate', data=df).fit()

# Print the model summary
print(model.summary())

# Extract the p-value and R-squared score
p_value = model.pvalues['Attendance_Rate']
r_squared = model.rsquared

# Print the p-value and R-squared score
print(f"P-value: {p_value:.4f}")
print(f"R-squared: {r_squared:.4f}")
```
